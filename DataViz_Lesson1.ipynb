```python
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
```

```python
data = pd.read_csv('https://gist.githubusercontent.com/andriy-gazin/ba80540e622b6e01b098ec623af236b1/raw/93be913d9891ec8b71ddce15ae81fb7321a063b7/wbdata.csv')

data.head()
```

```python
# data.plot.scatter(x="pop", y="gdpPerCap")
data1=data.filter(items=["continent", "gdpPerCap"])
# data1.groupby(['continent']).sum()
data2=data1.groupby('continent').sum()
data2.head()
data2.plot.pie(x="continent", y="gdpPerCap", figsize=(10, 10))
plt.show()
```
