```python
import altair as alt
import pandas as pd

from altair import datum
from vega_datasets import data
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from functools import partial
```

```python
df = pd.read_csv('scrubbed_UFO.csv')

test = df.dropna(subset=['latitude', 'longitude ']).copy()


test.count
```

```python


df1 = test.iloc[:1000]
df1 = df1.rename(columns={'longitude ': 'longitude'})
df1.replace(to_replace='&#39', value='\'', inplace=True, regex=True)
df1.replace(to_replace='&#44', value=',', inplace=True, regex=True)

df1['location'] = df1['latitude'].astype(str) + ' ' + df1['longitude'].astype(str)

geolocator = Nominatim(user_agent="final_project")
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)
df1['address_geocoded'] = df1['location'].apply(partial(reverse, language='en'))
df1['country_geocoded'] = df1['address_geocoded'].apply(lambda add: add.raw['address']['country'])

df1.head()
```
