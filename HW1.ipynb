```python
import altair as alt
import pandas as pd

# for the notebook only (not for JupyterLab) run this command once per session
alt.renderers.enable('notebook')
```

```python
df = pd.read_csv('population_prospects.csv')
df.head()
```

```python
estimatePlot = alt.Chart(df).mark_line().encode(
    x = alt.X('year:O', axis=alt.Axis(values=[1950, 1991, 2020, 2100], labelPadding=15, labelAlign="center")),
    y = alt.Y('population:Q',  scale = alt.Scale(domain=[15000, 55000])),
    color = alt.value('#3288bd')
).transform_filter(
'datum.type == "estimate"'
)

momentumPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('population:Q'),
    color = alt.value('#d53e4f'),
).transform_filter(
'datum.type == "momentum"'
)

instRepPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('population:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "instant replacement"'
)

highVarPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('population:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "high variant"'
)

medVarPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('population:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "medium variant"'
)

zerMigPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('population:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "zero migration"'
)

constFertPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('population:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "constant fertility"'
)

constMortPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('population:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "constant mortality"'
)

noChangePlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('population:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "no change"'
)

lowVarPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('population:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "low variant"'
)

alt.layer(
estimatePlot,
momentumPlot,
instRepPlot,
highVarPlot,
medVarPlot,
zerMigPlot,
constFertPlot,
constMortPlot,
noChangePlot,
lowVarPlot
).configure(background='#F3F7F7'
).configure_axis(grid=True, gridColor='#5D646F', gridDash=[1,2], gridOpacity=0.5
).configure_header(titleColor='#5D646F', titleFontSize=30
).properties(width=700, height=400)

```
