```python
import altair as alt
import pandas as pd

# for the notebook only (not for JupyterLab) run this command once per session
alt.renderers.enable('notebook')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/kai-wren/UCU_DataViz/master/population_prospects.csv')
df.head()
```

```python
estimatePlot = alt.Chart(df).mark_line().encode(
    x = alt.X('year:O', axis=alt.Axis(values=[1950, 1991, 2020, 2100], tickSize=10, labelPadding=5, labelAlign="center"),
             title='Data: UN World Population Prospects'),
    y = alt.Y('pop:Q',  axis=alt.Axis(values=[15, 20, 25, 30, 35, 40, 45, 50], tickSize=6),  
              scale = alt.Scale(domain=[15, 52]), title='Total population, million'),
    color = alt.value('#3288bd')
).transform_filter(
'datum.type == "estimate"'
).transform_calculate(
pop='datum.population/1000'
)

momentumPlot = alt.Chart(df).mark_line(strokeDash=[4, 1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('pop:Q'),
    color = alt.value('#d53e4f'),
).transform_filter(
'datum.type == "momentum"'
).transform_calculate(
pop='datum.population/1000'
)

instRepPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('pop:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "instant replacement"'
).transform_calculate(
pop='datum.population/1000'
)

highVarPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('pop:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "high variant"'
).transform_calculate(
pop='datum.population/1000'
)

medVarPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('pop:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "medium variant"'
).transform_calculate(
pop='datum.population/1000'
)

zerMigPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('pop:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "zero migration"'
).transform_calculate(
pop='datum.population/1000'
)

constFertPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('pop:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "constant fertility"'
).transform_calculate(
pop='datum.population/1000'
)

constMortPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('pop:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "constant mortality"'
).transform_calculate(
pop='datum.population/1000'
)

noChangePlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('pop:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "no change"'
).transform_calculate(
pop='datum.population/1000'
)

lowVarPlot = alt.Chart(df).mark_line(strokeDash=[4,1]).encode(
    x = alt.X('year:O'),
    y = alt.Y('pop:Q'),
    color = alt.value('#d53e4f')
).transform_filter(
'datum.type == "low variant"'
).transform_calculate(
pop='datum.population/1000'
)

textEstimate = alt.Chart(df).mark_text(text='estimate', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#3288bd'),
x=alt.value(180),
y=alt.value(20)
)

textProjection = alt.Chart(df).mark_text(text='projection', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#d53e4f'),
x=alt.value(550),
y=alt.value(20)
)

textInstRep = alt.Chart(df).mark_text(text='instant replacement', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#5D646F'),
x=alt.value(820),
y=alt.value(160)
)

textHighVar = alt.Chart(df).mark_text(text='high variant', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#d53e4f'),
x=alt.value(798),
y=alt.value(175)
)

textMomentum = alt.Chart(df).mark_text(text='momentum', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#5D646F'),
x=alt.value(785),
y=alt.value(210)
)

textMedVar = alt.Chart(df).mark_text(text='medium variant', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#d53e4f'),
x=alt.value(805),
y=alt.value(300)
)

textZerMig = alt.Chart(df).mark_text(text='zero migration', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#5D646F'),
x=alt.value(805),
y=alt.value(311)
)

textConstFert = alt.Chart(df).mark_text(text='constant fertility', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#5D646F'),
x=alt.value(818),
y=alt.value(335)
)

textConstMort = alt.Chart(df).mark_text(text='constant mortality', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#5D646F'),
x=alt.value(818),
y=alt.value(345)
)

textNoChange = alt.Chart(df).mark_text(text='no change', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#5D646F'),
x=alt.value(790),
y=alt.value(375)
)

textLowVar = alt.Chart(df).mark_text(text='low variant', fontSize=13, fontWeight=100, font='Ubuntu Mono').encode(
color=alt.value('#d53e4f'),
x=alt.value(795),
y=alt.value(400)
)

alt.layer(
estimatePlot,
momentumPlot,
instRepPlot,
highVarPlot,
medVarPlot,
zerMigPlot,
constFertPlot,
constMortPlot,
noChangePlot,
lowVarPlot,
textEstimate,
textProjection,
textInstRep,
textHighVar,
textMomentum,
textMedVar,
textZerMig,
textConstFert,
textConstMort,
textNoChange,
textLowVar
).configure(background='#F3F7F7', title=alt.TitleConfig(anchor='start', color='#5D646F', 
                                                        fontSize=30, font='Ubuntu Mono', fontWeight=900), padding=40
).configure_axis(grid=True, gridColor='#5D646F', gridDash=[1, 0, 0.5], gridOpacity=0.3, 
                 tickColor='#5D646F', tickDash=[1, 0, 0.5], tickOpacity=0.3,  
                 domain=False, labelAngle=0, labelFont='Ubuntu Mono', labelFontSize=13, labelColor='#5D646F', 
                 titleFont='Ubuntu Mono', titleColor='#5D646F', titleFontWeight=300
).configure_view(stroke='#F3F7F7'
).configure_axisY(
    titleFontSize=18,
    titleAngle=0,
    titleY=-10,
    titleX=87,
).configure_axisX(
    titleY=60,
    titleX=765,
).properties(width=750, height=400, title='Ukraine Population Prospects')

```
