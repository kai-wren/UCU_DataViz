```
import pandas as pd
import altair as alt
```

```
data = alt.Data(
    values = [{"interval":"100-119","share":18.3,"region_type":"rural"},
              {"interval":"120-139","share":32.7,"region_type":"rural"},
              {"interval":"140-159","share":28.8,"region_type":"rural"},
              {"interval":"160-179","share":16.7,"region_type":"rural"},
              {"interval":"180-200","share":3.5,"region_type":"rural"},
              {"interval":"100-119","share":9.2,"region_type":"urban"},
              {"interval":"120-139","share":22.2,"region_type":"urban"},
              {"interval":"140-159","share":28.2,"region_type":"urban"},
              {"interval":"160-179","share":27.8,"region_type":"urban"},
              {"interval":"180-200","share":12.6,"region_type":"urban"}])
```

```
alt.Chart(data).mark_bar().encode(
    x = alt.X('share:Q'),
    y = alt.Y('region_type:N'),
    color = alt.Color('interval:N')
)
```

```
base = alt.Chart(data).transform_calculate(
    diff='(datum.share)'
)

marks = base.mark_point().encode(
    y=alt.Y('share:Q', title='Share of marks'),
    x=alt.X('interval:N', title='Marks intervals'),
    shape = alt.Shape('region_type:O', title='Type of region'),
    color = alt.Color('region_type:O', scale=alt.Scale(scheme='plasma'))
)

line = base.mark_line().encode(
    y = alt.Y('min(share):Q'),
    x=alt.X('interval:N'),
    y2=alt.Y2('max(share):Q')
)

# text = base.mark_text(yOffset = -10).encode(
#     x = alt.X('mean(share):Q'),
#     y=alt.Y('interval:N'),
#     text=alt.Text('diff:Q')
# )

(marks+line).configure_legend(
    orient='right'
).configure_title(font = 'Arial Bold', fontSize = 18).properties(width=685, height=300, 
                                                                 title = alt.TitleParams(orient='top', anchor='start', text = 'Share of Grades per Interval', 
                                                                                         subtitle = 'Difference between rural and urban areas',
                                                                                        font='Arial Bold', fontSize=18, color = '#000000',
                                                                                         subtitleColor='#000000', subtitleFontSize=14, subtitleFont = 'Arial'
)
).configure_axisY(
    titleAngle=0, titleAnchor='start', titleY=-15, titleX=-18, grid=True, gridColor = '#DEDDDD', gridWidth = 1, domain = False
).configure_axis(
    domain = True, ticks = False, grid=False
).configure_axisX(
    labelAngle = 0
)
```
